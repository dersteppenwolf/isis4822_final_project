## Interquartile Ranges & Outliers

https://www.purplemath.com/modules/boxwhisk3.htm


That is, if a data point is below Q1 – 1.5×IQR or above Q3 + 1.5×IQR, it is viewed as being too far from the central values to be reasonable. Maybe you bumped the weigh-scale when you were making that one measurement, or maybe your lab partner is an idiot and you should never have let him touch any of the equipment. Who knows? But whatever their cause, the outliers are those points that don't seem to "fit".

```python
# Computing IQR
Q1 = df['costo_procedimiento'].quantile(0.25)
Q3 = df['costo_procedimiento'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
filtered = df.query('(@Q1 - 1.5 * @IQR) <= costo_procedimiento <= (@Q3 + 1.5 * @IQR)')
filtered.info()
```

```python
ax = sns.boxplot(x=filtered["costo_procedimiento"])
```

# Procesar códigos

```python
df = pd.read_csv('costos_20181122_depurado_outliers_filtered.tsv', sep = '\t')
df = df.rename(columns={ df.columns[0]: "idx" })
df.head(2)
```

# codificar régimen y nivel sisben

```python
df_regimen_list = df.groupby(['regimen_de_administradora'], as_index=False)['costo_procedimiento', 'numero_personas_atendidas'].agg('sum')
df_regimen_list["costo_proc_usuario"] = df_regimen_list["costo_procedimiento"] / df_regimen_list["numero_personas_atendidas"] 
df_regimen_list['code_regimen'] = df_regimen_list.index + 1
df_regimen_list.head(5)
```

```python
df = pd.merge(df, df_regimen_list[['regimen_de_administradora','code_regimen']], on='regimen_de_administradora', how='outer')
df.head(2)
```

```python
df_regimen_list = df_regimen_list.sort_values(by=['regimen_de_administradora'])
df_regimen_list = df_regimen_list.rename(columns={ "regimen_de_administradora": "label" })
df_regimen_list = df_regimen_list.rename(columns={ "code_regimen": "code" })
df_regimen_list.to_csv('domain_regimen.tsv', sep = '\t', index=False)
df_regimen_list.head(5)
```

# Convertir strings a código - label

```python
def custom_split(s):
    #print(s)
    parts = s.strip().split("-")
    code = parts[0].strip()
    label = "-".join( parts[1:] ).replace('"', '').strip()
    #print(parts) 
    return  pd.Series( [code  , label]  )

#print(custom_split("EPS003 - CAFESALUD - Colombia") )
print(custom_split(' EPS003 - CAFESALUD-Colombia "al menos" a - b . c - d ') )
```

```python
df.head(2)
```

```python
df[['code_admin','label_admin']] = df['administradora'].apply(custom_split)
df[['code_depto','label_depto']] = df['departamento'].apply(custom_split)
df[['code_prestador','label_prestador']] = df['prestador'].apply(custom_split)
df[['code_procedimiento','label_procedimiento']] = df['procedimiento'].apply(custom_split)
```
