# remover no reportado / no definidos

```python
df = df[df['regimen_de_administradora'] != 'NO DEFINIDO']
df = df[df['regimen_de_administradora'] != 'NA : NO APLICA']
df = df[df['prestador'] != '0 - NO DEFINIDO']
df = df[df['departamento'] != '-1 - NO DEFINIDO']


df = df[df['seccion'] != '1 - NO DEFINIDO']
df = df[df['seccion'] != '2 - NO APLICA']
df = df[df['seccion'] != '3 - NO REPORTADO']
df = df[df['seccion'] != 'NA - ANEXOS']

df = df[df['capitulo'] != '1 - NO DEFINIDO']
df = df[df['procedimiento'] != '-1 - NO DEFINIDO']
df = df[df['procedimiento'] != '1 - NO DEFINIDO	']



```

```python
df = df.drop('valor_consulta', 1)
df = df.drop('valor_cuota_moderadora_-_consulta', 1)
df = df.drop('neto_a_pagar_-_consulta', 1)
```

```python
df['costo_procedimiento'] = df['costo_procedimiento'].str.replace(".", "")
```

```python
# imprime filas con errores
# print (df[pd.to_numeric(df['costo_procedimiento'], errors='coerce').isnull()]  ['costo_procedimiento']  )
```

```python
df['costo_procedimiento'] = df['costo_procedimiento'].apply(pd.to_numeric, errors='coerce')
#
#df["costo_procedimiento"] = df["costo_procedimiento"].astype(float)
```

```python
df.info()
```

```python
df = df[pd.notnull(df['costo_procedimiento'])]
df.info()
```

# procedimientos - solo primer nivel  seccion

```python
df_seccion = df.copy()
columns = ['capitulo', 'grupo', 'subgrupo', 'categoria', 'procedimiento'  ]
df_seccion.drop(columns,inplace=True ,  axis=1)
df_seccion.info()
```

```python
df_seccion.to_csv('costos_admin_prest_seccionproc_depurado.tsv', sep = '\t')
```

```python
df_seccion[['code_depto','label_depto']] = df_seccion['departamento'].str.split('-',expand=True) 
```

```python
#df_seccion[['code_admin','label_admin']] = df_seccion['administradora'].str.split('-',expand=True) 
#df_seccion[['code_prestador','label_prestador']] = df_seccion['prestador'].str.split('-',expand=True) 
df_seccion[['code_seccion','label_seccion']] = df_seccion['seccion'].str.split('-',expand=True) 
```

```python
df_seccion.head(2)
```

```python
df_seccion_codes = df_seccion.copy()
columns = ['departamento', 'seccion', 'label_depto', 'label_seccion'  ]
df_seccion_codes.drop(columns,inplace=True ,  axis=1)
df_seccion_codes.head(2)
```

```python
df_seccion_codes.to_csv('costos_admin_prest_seccionproc_depurado_codes.tsv', sep = '\t')
```

# exportar secciones

```python
df_seccion_list = df_seccion.groupby(['code_seccion','label_seccion'], as_index=False)['costo_procedimiento', 'numero_personas_atendidas'].agg('sum') 
# conteo_de_administradoras	conteo_de_prestadores		numero_de_atenciones	

df_seccion_list["costo_proc_usuario"] = df_seccion_list["costo_procedimiento"] / df_seccion_list["numero_personas_atendidas"] 


df_seccion_list = df_seccion_list.sort_values(by=['label_seccion'])
df_seccion_list.head(10)
```

```python
df_seccion_list.to_csv('costos_secciones.tsv', sep = '\t', index=False)
```

# exportar departamentos

```python
df_seccion_depto = df_seccion.groupby(['departamento'], as_index=False)['costo_procedimiento', 'numero_personas_atendidas'].agg('sum') 
# conteo_de_administradoras	conteo_de_prestadores		numero_de_atenciones	

df_seccion_depto["costo_proc_usuario"] = df_seccion_depto["costo_procedimiento"] / df_seccion_depto["numero_personas_atendidas"] 

df_seccion_depto[['code','label']] = df_seccion_depto['departamento'].str.split('-',expand=True) 
df_seccion_depto = df_seccion_depto.sort_values(by=['label'])
df_seccion_depto.head(32)
```

```python
df_seccion_depto.to_csv('costos_departamentos.tsv', sep = '\t', index=False)
```

# exportar años 

```python
df_anio = df_seccion.groupby(['anno'], as_index=False)['costo_procedimiento', 'numero_personas_atendidas'].agg('sum') 
# conteo_de_administradoras	conteo_de_prestadores		numero_de_atenciones	

df_anio["costo_proc_usuario"] = df_anio["costo_procedimiento"] / df_anio["numero_personas_atendidas"] 
df_anio["label"] = df_anio["anno"] 
#df_seccion_depto[['code','label']] = df_seccion_depto['departamento'].str.split('-',expand=True) 
df_anio = df_anio.sort_values(by=['label'])
df_anio.head(32)
```

```python
df_anio.to_csv('costos_anios.tsv', sep = '\t', index=False)
```
