```python
%matplotlib inline

import matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_profiling
import seaborn as sns
import unidecode
from pandas_profiling.describe import describe
```

```python
df = pd.read_csv('procedimientos.csv' )
df.head()
```

```python
df  = df.rename(columns=lambda x: unidecode.unidecode(x.strip().replace(' ', '_').lower()))
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum() 
```

```python
df.isnull().sum() / len(df) * 100
```

```python
list(df)
```

```python
# tale of nulls 
# https://towardsdatascience.com/the-tale-of-missing-values-in-python-c96beb0e8a9d
df.info()
```

```python
df = df.drop('valor_consulta', 1)
df = df.drop('valor_cuota_moderadora_-_consulta', 1)
df = df.drop('neto_a_pagar_-_consulta', 1)

```

```python
df = df[pd.notnull(df['costo_procedimiento'])]
df.info()
```

```python
df['costo_procedimiento'] = df['costo_procedimiento'].str.replace(".", "")
# imprime filas con errores
print (df[pd.to_numeric(df['costo_procedimiento'], errors='coerce').isnull()]  ['costo_procedimiento']  )
```

```python
df['costo_procedimiento'] = df['costo_procedimiento'].apply(pd.to_numeric, errors='coerce')
#
#df["costo_procedimiento"] = df["costo_procedimiento"].astype(float)
df.info()
```

```python
df.head()
```

```python
pandas_profiling.ProfileReport(df)
# describe(df, check_correlation=False)
```

# Remover outliers 

```python
'''
https://machinelearningmastery.com/how-to-use-statistics-to-identify-outliers-in-data/

Minimum	1
5-th percentile	4371.1
Q1	30554
Median	96397
Q3	357600
95-th percentile	2457100
Maximum	103009000000
Range	103008999999
Interquartile range	327050

'''

q = df["costo_procedimiento"].quantile(0.99)
q
```

```python
df = df[df["costo_procedimiento"] < q]
df.info()
```

```python
pandas_profiling.ProfileReport(df)
```

```python

df_top_freq =  df.groupby(['anno', 'procedimiento'], as_index=False).sum().sort_values(
    ['anno' , "costo_procedimiento"],  ascending=False  
    ).reset_index()

df_top_freq = df_top_freq.sort_values(['anno', 'costo_procedimiento'], ascending=[True, False])
```

```python
df_top_freq.head(5)
```

# top 5  procedimientos por año  / costo

```python
top_n = 5
df_top_n = df_top_freq.groupby('anno', as_index=False).apply(lambda x: x.nlargest(top_n, 'costo_procedimiento'))
df_top_n.head(10)
```

```python
df_top_n.to_csv('procedimientos_depurado_top_5_anno.tsv', sep = '\t')
```
