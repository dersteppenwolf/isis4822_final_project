# remover outliers


```python
q = df["costo_procedimiento"].quantile(0.99)
q
```

```python
df = df[df["costo_procedimiento"] < q]
df.info()
```

```python
pandas_profiling.ProfileReport(df)
```

# remover no reportado / no definidos

```python

df = df[df['sexo'] != 'NR - NO REPORTADO']
df = df[df['sexo'] != 'NO DEFINIDO']
df = df[df['departamento'] != '-1 - NO DEFINIDO']
df = df[df['regimen_de_administradora'] != 'NO DEFINIDO']
df = df[df['zona'] != 'NO DEFINIDO']
df = df[df['tipo_de_usuario'] != '0 - NO DEFINIDO']


df.info()
```

```python
df.to_csv('costos_depurado.tsv', sep = '\t')
```
